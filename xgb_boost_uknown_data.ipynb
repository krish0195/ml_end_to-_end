```python
#label encoding of tgt varaible
tgt_map = {    'Flexion_football':0,
               'Flexion_relaxed':1,
               'Flexion_stiffed':2,
               'Lateral_football':3,
               'Lateral_relaxed':4,
               'Lateral_stiffed':5
              }
df_fulldata_train["Direction"]=df_fulldata_train.Direction.map(tgt_map)
df_fulldata_train=df_fulldata_train.drop(columns=["time","Subject","Sr No"],axis=1)

df_fulldata_test["Direction"]=df_fulldata_test.Direction.map(tgt_map)
df_fulldata_test=df_fulldata_test.drop(columns=["time","Subject","Sr No"],axis=1)


x=df_fulldata_train.drop(columns=['Direction'])
y=df_fulldata_train['Direction']



x_unknown_test=df_fulldata_test.drop(columns=['Direction'])
y_unknown_test=df_fulldata_test['Direction']




from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()


scalar.fit(x)
x_scaled = pd.DataFrame(scalar.transform(x),columns=x.columns)


scalar.fit(x_unknown_test)
x_unknown_scaled = pd.DataFrame(scalar.transform(x_unknown_test),columns=x_unknown_test.columns)
```

```python
print(x_scaled.shape)
print(x_unknown_scaled.shape)
```

```python
#validation data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.20, stratify = y, random_state = 43,shuffle=True)

print("x_train shape is",x_train.shape)
print("y_train shape is",y_train.shape)

print("x_test shape is",x_test.shape)
print("y_test shape is",y_test.shape)
```

```python
#Test data

print("x_test_unknown shape is",x_unknown_scaled.shape)
print("y_test shape is",y_unknown_test.shape)
```

```python
#Classification Metrics validation data:

def prediction(x_train,y_train,x_test,y_test,y_train_predict,y_test_predict):

  #Accuracy Check of Train
  train_accuracy = metrics.accuracy_score(y_train,y_train_predict)
  print("The Train Accuracy:",round(train_accuracy,3))

# Accuracy Check of Test
  test_accuracy = metrics.accuracy_score(y_test,y_test_predict)
  print("The Test Accuracy",round(test_accuracy,4))

  return train_accuracy,test_accuracy
```

```python
#4) XG Boost calssifier-Known validation data
import xgboost as xgb
xgb_classifier = xgb.XGBClassifier()

xgb_classifier.fit(x_train, y_train)
y_train_predict = xgb_classifier.predict(x_train)
y_test_predict = xgb_classifier.predict(x_test)

print("the Train and Test accuracy of XG Boost calssifier ")
XGB_train_accuracy, XGB_test_accuracy=prediction(x_train,y_train,x_test,y_test,y_train_predict,y_test_predict)
```

```python
#4) XG Boost calssifier-unKnown test data
import xgboost as xgb
xgb_classifier = xgb.XGBClassifier()

xgb_classifier.fit(x_train, y_train)

y_unkown_test_predict = xgb_classifier.predict(x_unknown_test)

print("the Test accuracy of XG Boost calssifier unknown data ")

# Accuracy Check of Test
test_accuracy_unknown = metrics.accuracy_score(y_unknown_test,y_unkown_test_predict)
print("The unknown Test Accuracy",round(test_accuracy_unknown,4))

```
